## Exercice 1 : Générer des coordonnées équidistantes
Nous voulons générer $n + 1$ coordonnées $x$ équidistantes dans $[a, b]$.

Stocker, pour `a = -2`; `b = 3` et `n= 20` les coordonnées $x$ dans une liste `xList`.

* **Q1.** Définir toutes les variables puis utiliser une boucle **for** et ajouter chaque coordonnée à la liste `xList` (*initialement vide*).

---
**Indications.**

Avec $n$ intervalles, correspondant à $n + 1$ points, dans $[a, b]$, chaque intervalle a une longueur $h = (b-a) / n$. Les coordonnées peuvent alors être générées par la formule `xi = a + i * h`; $i = 0,\dots, n$.

---


```python
n =20
a, b = -2, 3
h = (b - a) / n
xList = []
for i in range(n+1):
    xi = a + i * h
    xList.append(xi)

# vérification : Affiche xList
print(xList)
```

* **Q2.** Utiliser une liste de compréhension comme une implémentation alternative.

```python
xList = [a + i * h for i in range(n+1)]

# vérification : Affiche xList
print(xList)
print(type(xList))
```

* **Q3.** Vectoriser la liste résultante `xList` en un tableau numpy `xVect`. N’oubliez pas d’importer d’abord la fonction qui transforme les listes en tableaux à partir de numpy.

```python
from numpy import array
xVect = array(xList)

# vérification : Affiche xVect
print(xVect)
print(type(xVect))
```

* **Q4.** Les cyclo-harmoniques $
\left\{
\begin{array}{ll}
x(t) = \left( 1 + cos(\dfrac{p}{q}t)\right) \times cos(t) &\\
y(t) = \left( 1 + cos(\dfrac{p}{q}t)\right) \times sin(t)&
\end{array} 
\right.$ 
pour $p$ et $q$ entiers positifs et $ t\in[0, 2 q \pi]$.

```python
import matplotlib.pyplot as plt
import numpy as np

p = 8
q = 3

t = np.linspace(0, 2 * q * np.pi, 400)
# Cyclo-harmonique
tmp = 1 + np.cos(p / q * t)
x = tmp * np.cos(t)
y = tmp * np.sin(t)
plt.plot(x, y, 'r')
plt.show()
```

## Exercice 3 : Implémenter une fonction gaussienne


**Q1.** Créer la fonction: `gauss(x, m = 0, s = 1)`, qui modélise la gaussienne:
$$
\begin{equation}
f(x) =
{1\over\sqrt{2\pi }\, s}
\exp{\left[-\frac{1}{2}\left({x-m\over s}\right)^2\right]}
\label{_auto1} \tag{1}
\end{equation}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
def gauss(x, m = 0, s = 1):
    A = 1/np.sqrt(2*np.pi*s)
    B = -0.5 *((x-m)/s)**2
    return A*np.exp(B)
```

## Exercice 4 : La suite logistique
Soit $r$ un réel de $[0,4]$. On considère la suite logistique définie par
$$
u_{n+1} = r \times u_n(1 - u_n) \quad et \ u_0 = 0.5
$$
Cette suite très célèbre permet de modéliser certaines dynamiques de population. Elle a uncomportement très différent selon les valeurs de $r$. On se propose de tracer le nuage de pointssuivants.

**Q1.** Écrire une fonction `Liste_suite(u0,r,n)` qui renvoie la liste desnpremiers termes dela suite.

```python
def Liste_suite(u0,r,n):
    U = [0]*(n+1)
    U[0] = u0
    for i in range(0, n):
        U[i+1]= r*U[i]*(1-U[i])
       
    return U
```

**Q2.** Créer une subdivision uniforme `R` de l’intervalle $[0,4]$ en 101 points

```python
h = 4/100
R = []
for i in range(100):
    R.append(i*h)
```

**Q3.** Pour chaque élément `r` de `R`, représenter les points de coordonnées $(r,u_{100}),(r,u_{101}),\dots,(r,u_{200})$.

```python
u0 = 0.5
import matplotlib.pyplot as plt
for n in range(100, 201):
    for r in R:
        plt.plot(r, Liste_suite(u0,r,n)[-1], 'ro', ms =2)
        
plt.title("Diagramme de bifurcation")
plt.xlabel("r")
plt.ylabel(r"$U_n$")
plt.show()
```
